#**Handling Missing Values**

```python
Q1 = df_house["Sale Price"].quantile(0.25)
Q3 = df_house["Sale Price"].quantile(0.75)
IQR = Q3-Q1
lower_bound = Q1-1.5*IQR
upper_bound = Q3+1.5*IQR
outliers = df_house[(df_house["Sale Price"]<lower_bound) | (df_house["Sale Price"]>upper_bound)]
len(outliers)
```

```python
median_sale_price = df_house["Sale Price"].median()
df_house["Sale Price"] = df_house["Sale Price"].fillna(median_sale_price)
```

```python
Q1 = df_house["No of Bathrooms"].quantile(0.25)
Q3 = df_house["No of Bathrooms"].quantile(0.75)
IQR = Q3-Q1
lower_bound = Q1-1.5*IQR
upper_bound = Q3+1.5*IQR
outliers = df_house[(df_house["No of Bathrooms"]<lower_bound) | (df_house["No of Bathrooms"]>upper_bound)]
len(outliers)
```
